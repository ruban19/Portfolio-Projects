# We can also find an another pattern on product_name column that the last word of the product_name column values consist of the actual_product name and so lets slice that put it in a different column as it will be useful for our future analysis

```python
lst = list(df['product_name'].str.split())
lst
```

```python
product = []
for i in lst:
    x = i[-1]
    product.append(x)
product  
```

```python
df.insert(2,'actual_product',product)
```

```python
df['actual_product'].value_counts()
```

```python
products = []
for i in df['actual_product']:
    if i == 'shoes':
        products.append('Shoes')
    elif i == 'Flops':  
       products.append('Flip-Flops')
    elif i == 'flops':  
       products.append('Flip-Flops')
    elif i == 'Flop':  
       products.append('Flip-Flops')      
    elif i == 'Sandal':  
       products.append('Sandals')
    elif i == 'Sandal(Chappal)':  
       products.append('Sandals')
    elif i == 'Bags':  
       products.append('Sling Bags')
    else :
       products.append(i)
```

```python
df['actual_product'] = products  
```

# T- Test  
# 1) One sample T -Test

# Lets check one sample t - test for bounce_rate as it is a continuous value column

```python
df['bounce_rate'].mean()
```

```python
random = np.random.choice(df['bounce_rate'],30)
random
```

```python
np.mean(random)
```

```python
from scipy.stats import ttest_1samp
import scipy.stats as stats
```

```python
ttest,p1_value = ttest_1samp(random,0.504)
```

```python
p1_value
```

```python
if p1_value < 0.05:
    print('We reject the null hypothesis')
else:
    print('Null Hypothesis is True')
```

# 2) Two sample T- Test 

# Lets check two sample t - test for the above dataset

```python
df['buy_to_detail_rate'].mean()
```

```python
df['cart_to_detail_rate'].mean()
```

```python
random1 = np.random.choice(df['buy_to_detail_rate'],30)
random2 = np.random.choice(df['cart_to_detail_rate'],30)
random1
```

```python
random2
```

```python
ttest,p2_value = stats.ttest_ind(random1,random2,equal_var=False)
```

```python
p2_value
```

```python
if p2_value < 0.05:
    print('We reject the null hypothesis')
else:
    print('Null Hypothesis is True')
```
