```python
def cleaning(x):
    if x == "yearOfRegistration":
        x = "registration_year"
    elif x == "monthOfRegistration":
        x = "registration_month"
    elif x == "notRepairedDamage":
        x = "unrepaired_damage"
    elif  x == "dateCreated":
        x = "ad_created"   
    else:
        x = ''.join(['_'+i.lower() if i.isupper() else i for i in x]) 
    return x
cleaned_columns = []
for c in columns_copy:
    clean_column = cleaning(c)
    cleaned_columns.append(clean_column)
    
```
